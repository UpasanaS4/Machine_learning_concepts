## K Means Clustering with 

  
  
- random data points    


- scikit learn iris dataset  

- https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html  



```python
## Importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn

## Importing the model 

from sklearn.cluster import KMeans

## Inbuilt dataset

from sklearn import datasets
```

```python
## Basic plotting six random points

from matplotlib import style
style.use("ggplot")

X = [1, 5, 1.5, 8, 1, 9]
Y = [2, 8, 1.8, 8, 0.6, 11]

plt.scatter(X,Y)
plt.show()
```

```python
## Defining an array of random points for clustering

x = np.array([[1,2],
              [5,8],
              [1.5,1.8],
              [8,8],
              [1,0.6],
              [9,11],
             ])
```

```python
## Initializing k

kmeans = KMeans(n_clusters=2)
kmeans

```

```python
## Fitting the model

kmeans.fit(x)
```

```python
## Checking the centroids

centroids = kmeans.cluster_centers_
centroids
```

```python
## Checking the labels

labels = kmeans.labels_
labels
```

```python
## Printing both together

print("------Labels----\n")
print(labels,"\n")
print("-------Centroids ---\n")
print(centroids, "\n")
```

```python
## Defining two colors for two labels

colors = ["g.","r."]
```

```python
## Plotting with two actual clusters

colors = ["g.","r."]

## i is ranging from 0 to 6 i.e. 0,1,2,3,4,5

for i in range(len(x)):
    print("coordinate:", x[i], "label:", labels[i])
    plt.plot(x[i][0],x[i][1], colors[labels[i]], markersize = 10)
    
plt.scatter(centroids[:,0],centroids[:,1], marker = "x", s=150)
plt.show()
```

## Lets change the cluster numbers to k = 3

```python
## Initializing k

kmeans = KMeans(n_clusters=3)
kmeans

## Fitting the model

kmeans.fit(x)

## Getting updated centroids
centroids = kmeans.cluster_centers_


for i in range(len(x)):
    print("coordinate:", x[i], "label:", labels[i])
    plt.plot(x[i][0],x[i][1], colors[labels[i]], markersize = 10)
    
plt.scatter(centroids[:,0],centroids[:,1], marker = "x", s=150)
plt.show()
```

## What happens when k = 4?

```python
## Initializing k

kmeans = KMeans(n_clusters=4)
kmeans

## Fitting the model

kmeans.fit(x)

## Getting updated centroids
centroids = kmeans.cluster_centers_


for i in range(len(x)):
    print("coordinate:", x[i], "label:", labels[i])
    plt.plot(x[i][0],x[i][1], colors[labels[i]], markersize = 10)
    
plt.scatter(centroids[:,0],centroids[:,1], marker = "x", s=150)
plt.show()
```

# Working with inbuilt dataset "Iris" from sklearn

```python
## Loading the inbuilt dataset from iris

iris = datasets.load_iris()
```

```python
## Just taking the required array of data points 

X = iris["data"]
```

```python
## Three labels of the data

iris.target   ## iris["target"]
```

```python
## Converting array into a dataframe

df = pd.DataFrame(X)
```

```python
df.head()
```

```python
## Adding the column/feature name to the dataframe

df.columns = iris["feature_names"]
```

```python
df.head(2)
```

```python
## Adding a new column to df 

df["target"] = iris["target"]
```

```python
df.head(2)
```

```python
## Checking the target column 

df["target"].value_counts()
```

## Lets dump this data into scikit learn KMeans

```python
## Creating an object with my kmeans parameters

km = KMeans(n_clusters=3, max_iter=1000)  ## By default it is 300
km
```

```python
## Fitting the model

km.fit(X)
```

```python
## Checking the centroids

km.cluster_centers_
```

```python
## Checking the predicted labels

km.labels_
```

```python
## Checking the actual labels

iris.target
```

```python
## Creating a new column with the predicted labels

df["K Mean predicted label"] = km.labels_
```

```python
## Lets check the newly created column

df
```
